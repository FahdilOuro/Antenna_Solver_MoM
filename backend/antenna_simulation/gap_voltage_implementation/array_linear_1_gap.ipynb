```python
from backend.utils.gmsh_function import *
from backend.src.radiation_algorithm.radiation_algorithm import *
from backend.utils.file_path import *
```

```python
name = "array_2_strip"
save_mesh_folder = 'data/gmsh_files/'

path = setup_save_file_paths(name, save_mesh_folder)
```

```python
mm = 1e-3
width = 20  * mm
length = 2000 * mm
distance_between_strips = 2000 * mm

initial_mesh_size = 0.1
print(f"initial_mesh_size = {initial_mesh_size}")

feed_point = np.array([[-1, 0, 0],
                      [1, 0, 0]])

print(feed_point.shape[0])
```

```python
light_speed = 3e8

frequency = 75e6
wavelength = light_speed / frequency
print(f"wavelength = {wavelength} Meter")
```

```python
gmsh.initialize()
model_name  = "strip_antenna"

print(f"Longueur: {width} m, Hauteur: {length} m")

# Model creation
gmsh.model.add(model_name)
setup_performance_config()

# create 2 strips spaced by distance_between_strips
strip1 = gmsh.model.occ.addRectangle(-width/2 - distance_between_strips/2, -length/2, 0, width, length)
strip2 = gmsh.model.occ.addRectangle(-width/2 + distance_between_strips/2, -length/2, 0, width, length)


# synchronize after embedding
gmsh.model.occ.synchronize()

generate_and_save_mesh(path.geo, path.msh, initial_mesh_size)

# gmsh.fltk.run()

gmsh.finalize()
```

```python
extract_msh_to_mat(path.msh, path.mat)
```

```python
matrice_z, voltage, current, surface_current_density = radiation_algorithm(path.mat, frequency, feed_point, excitation_unit_vector='y', gap_width=3*initial_mesh_size/2)
```
