# create filename

```python
name = "plate"
save_mesh_folder = 'data/gmsh_files/'

# Define the path to save the file
path = setup_save_file_paths(name, save_mesh_folder)
```

```python
length = 2
width = 1

light_speed = 3e8
frequency = light_speed / (4 * length)
print(f"frequency = {frequency * 1e-06} MHz")
wavelength = light_speed / frequency
print(f"wavelength = {wavelength}")

initial_mesh_size = wavelength / 10
print(f"initial_mesh_size = {initial_mesh_size}")
```

```python
# Initialize Gmsh
gmsh.initialize()
gmsh.model.add("plate_gmsh")
setup_performance_config()

# Use Open Cascade to create a rectangle (x, y, z, width, width)
plate = gmsh.model.occ.addRectangle(0, 0, 0, length, width)
gmsh.model.occ.synchronize()

generate_and_save_mesh(path.geo, path.msh, initial_mesh_size)

# Close Gmsh
gmsh.finalize()
```

```python
wave_incident_direction = np.array([0, 0, -1])
polarization=np.array([0, 1, 0])
```
