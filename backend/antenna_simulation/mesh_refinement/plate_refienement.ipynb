# create filename

```python
length = 2
width = 1
initial_mesh_size = length * 0.4
name = "plate"
save_mesh_folder = 'data/gmsh_files/'

# Define the path to save the file
path = setup_gmsh_file_paths(name, save_mesh_folder)
```

```python
# Initialize Gmsh
gmsh.initialize()
gmsh.model.add("plate_gmsh")

# Use Open Cascade to create a rectangle (x, y, z, width, width)
plate = gmsh.model.occ.addRectangle(0, 0, 0, length, width)
gmsh.model.occ.synchronize()

generate_and_save_mesh(path.geo, path.msh, initial_mesh_size)

# Close Gmsh
gmsh.finalize()
```

```python
spacing = 0.41
grid_points, radius_vicinity = generate_embedded_grid(path.geo, spacing)
```

```python
wave_incident_direction = np.array([0, 0, -1])
polarization=np.array([0, 1, 0])

extract_msh_to_mat(path.msh, path.mat)

light_speed = 3e8
frequency = light_speed / (4 * length)
print(f"frequency = {frequency * 1e-06} MHz")

settings, sizes = initialize_refinement_config(grid_points, initial_mesh_size)
```

# Refinement mesh

```python
run_refinement_cycle(frequency, wave_incident_direction, polarization, settings, sizes, grid_points, radius_vicinity, path)
```
