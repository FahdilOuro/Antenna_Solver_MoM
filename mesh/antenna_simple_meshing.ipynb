```python
import triangle as tr
import numpy as np
import plotly.figure_factory as ff
```

```python
def show_mesh(points, triangles):
    # Normalisation pour l'aspect ratio
    x_, y_, z_ = points
    fig_scale = max(max(x_) - min(x_), max(y_) - min(y_))
    x_scale = (max(x_) - min(x_)) / fig_scale
    y_scale = (max(y_) - min(y_)) / fig_scale
    z_scale = 0.3  # Z scale arbitraire pour une meilleure visualisation

    # Création de la figure 3D avec Plotly
    fig = ff.create_trisurf(
        x=x_,
        y=y_,
        z=z_,
        simplices=triangles[:3, :].T,
        color_func=list(range(len(triangles[:3, :].T))),  # Couleurs basées sur les indices des triangles
        show_colorbar=False,
        title="Maillage triangulaire",
        aspectratio=dict(x=x_scale, y=y_scale, z=z_scale)
    )
    # Afficher la figure
    fig.show()
```

```python
Lx = 0.02
Ly = 2
# Définition des sommets du contour (ex: un rectangle)
points = np.array([
    [0, 0], [Lx, 0], [Lx, Ly], [0, Ly]  # Rectangle 1x1
])

# Indices des segments formant le contour
segments = np.array([
    [0, 1], [1, 2], [2, 3], [3, 0]
])

# Définition du point central où l'on veut placer une arête
center = np.array([0.5, 0.5])  # Point cible

# Création d'une petite arête autour de ce point
if Lx >= Ly:
    delta = Lx/100
else:
    delta = Ly/100

p1 = center + np.array([-delta, 0])  # Point à gauche
p2 = center + np.array([delta, 0])   # Point à droite

print("p1 = ", p1)
print("p2 = ", p2)

# Ajouter ces points au maillage
extra_points = np.array([p1, p2])

# Indice des nouveaux points
index_p1 = len(points)
index_p2 = len(points) + 1

# Ajout du segment contraint entre p1 et p2
extra_segment = np.array([[index_p1, index_p2]])

# Construire le dictionnaire pour Triangle
plate = {
    'vertices': np.vstack((points, extra_points)),  # Ajout des points
    'segments': np.vstack((segments, extra_segment))  # Ajout du segment contraint
}
"""
plate = {
    'vertices': points,  # Ajout des points
    'segments': segments  # Ajout du segment contraint
}"""
```

```python
# Générer le maillage avec contraintes
plate_mesh = tr.triangulate(plate, 'pq30a0.01')

nbr_of_points = plate_mesh['vertices'].shape[0]
nbr_of_triangles = plate_mesh['triangles'].shape[0]
print(f"Nombre de points = {nbr_of_points}")
print(f"Nombre de triangles = {nbr_of_triangles}")

triangles = np.zeros((4, nbr_of_triangles), dtype=int)
triangles[:3, :] = plate_mesh['triangles'].T
points = np.zeros((3, nbr_of_points))
points[:2, :] = plate_mesh['vertices'].T

print(f"Matrice points shape = {points.shape}")
print(f"Matrice triangles shape = {triangles.shape}")

show_mesh(points, triangles)
```
