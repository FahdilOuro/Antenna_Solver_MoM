```python
from utils.ifa_meander_project_v2.geometry import *
from matplotlib import pyplot as plt
```

```python
a = 15 / 1000  # Largeur
b = 30 / 1000  # Hauteur

largeur_piste = 1.2 / 1000
distance_meandre = 7 / 1000
```

```python
print(f"old distance meandres {distance_meandre * 1000} mm")
print(f"old largeur de piste {largeur_piste * 1000} mm")

x, y, N, distance_meandre = ifa_creation(largeur=a, hauteur=b, width=largeur_piste, distance_meandre=distance_meandre)
x_m, y_m = trace_meander(x, y, largeur_piste)
```

```python
plt.figure(figsize=(7, 7))
plt.scatter(x_m*1000, y_m*1000, color='red', label='Points', s=50)
plt.plot(x_m*1000, y_m*1000, '-b', linewidth=2, label='Contour')

# Boucle pour ajouter les indices
for k in range(len(x_m)):
    plt.text(x_m[k]*1000, y_m[k]*1000, f' {k+1}', 
             verticalalignment='bottom', 
             horizontalalignment='left',
             fontsize=10,
             color='black')

plt.xlabel('x_m (mm)', fontsize=12)
plt.ylabel('y_m (mm)', fontsize=12)
plt.title("Graphique des points avec numéros d'ordre", fontsize=14)
plt.grid(True)
plt.axis('equal')
plt.legend(fontsize=10)
plt.show()
```
